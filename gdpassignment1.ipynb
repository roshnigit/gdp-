```python
C1D= data1bfc.loc[:,['Items','Gujarat','Karnataka','Maharashtra','Tamil Nadu','Uttar Pradesh']]

C2D= data1bfc.loc[:,['Items','Andhra Pradesh','Bihar','Haryana','Kerala','Madhya Pradesh','Odisha','Punjab','Rajasthan','Telangana']]

C3D= data1bfc.loc[:,['Items','Assam','Chhatisgarh','Goa','Himachal','Jammu and Kashmir','Jharkhand','Tripura','Uttrakhand']]

C4D= data1bfc.loc[:,['Items','Arunachal Pradesh','Manipur','Meghalaya','Mizoram','Nagaland','Sikkim']]

C4D['Total']=C4D.sum(axis=1)
C3D['Total']=C3D.sum(axis=1)
C2D['Total']=C2D.sum(axis=1)
C1D['Total']=C1D.sum(axis=1)
C1D.set_index('Items', inplace=True)
C2D.set_index('Items', inplace=True)
C3D.set_index('Items', inplace=True)
C4D.set_index('Items', inplace=True)
C1D.sort_values(by='Total',ascending= False)
C2D.sort_values(by='Total',ascending= False)
C3D.sort_values(by='Total',ascending= False)
C4D.sort_values(by='Total',ascending= False)
C1D

```

```python
dropout=pd.read_csv("rs_session243_au570_1.1.csv")
dropout1415=dropout.loc[:,['Level of Education - State','Primary - 2014-2015','Upper Primary - 2014-2015','Secondary - 2014-2015']]
dropstate= dropout1415.iloc[[1,2,3,4,6,10,11,12,13,14,15,16,17,19,20,21,22,23,24,25,27,28,29,30,31,32,33,34],:]
ds=dropstate.set_index('Level of Education - State')
```

```python
dfb= data1bf2.T
dfgc=dfb[32]
mf1=pd.concat([ds,dfgc],axis=1,sort= True)
mf2=mf1.drop('Items')
mf3=mf2.rename(index=str,columns={32:"Per Capita GDP"})
```

```python
mf3['Per Capita GDP'] = mf3['Per Capita GDP'].apply(np.float64)
mf3['log gdp']=np.log(mf3['Per Capita GDP'])
```

```python
mf3g=mf3.iloc[:,[0,1,2,4]]
```

```python
cur= mf3g.corr()
round(cur,4)
```

```python
# figure size
plt.figure(figsize=(10,8))

# heatmap
sns.heatmap(cur, cmap="YlGnBu", annot=True)
plt.show()
```

```python
mf3gs=mf3g.sort_values(by='log gdp',ascending= False)
mf3gs.plot(y=['Primary - 2014-2015','Upper Primary - 2014-2015','Secondary - 2014-2015','log gdp'], figsize=(20,10),kind="bar")
```
